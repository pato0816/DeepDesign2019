```
import tensorflow as tf
from tensorflow import keras

import numpy as mp
import matplotlib.pyplot as plt

print(tf.__version__)
```

Download the data


```
fashion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels),(test_images, test_labels) = fashion_mnist.load_data()
```

```
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle boot']
```

```
train_images = train_images/255.0
test_images = test_images/255.0
```

```
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize = (10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid('off')
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i]])
    
```

```
#HERE WE CREATE OUR MODEL //ALL THE WORK BEFORE IS JUST IMPORTING AND CHECKING THE DATASET
#deep learning, sequential is stacking the layers
#keras.sequential describes the architecture of the neural network
#128 neurons.. hyperparameter of number of functions (minifunctions /neurons that will trigger either 1 or 0)
#128 is the number of neurons on the first layer you can also experiment with this and kernel size, activation sequence, etc.
#dense 10 you can experiment with different numer of layers, however the final one should have 10 as there are 10 number classes
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(10,activation=tf.nn.softmax)
    
])
```

```
tf.optimizers.Adam()
```

```
#HERE WE CREATE THE MEASURES TO SEE HOW OUR MODEL IS PERFORMING
#measure how well our networrk is doing
#setting up the parameters that the network is going to learn from
#model.compile(optimizer=tf.train.AdamOptimizer(),
model.compile(optimizer=tf.optimizers.Adam(), loss="sparse_categorical_crossentropy",
              metrics=['accuracy'])
```

```
#HERE WE TRAIN OUR MODEL
model.fit(train_images,train_labels, epochs=5)
#the more epochs the more acurracy.... if you have a more complex model you might want more epochs but there is computation time
```

```
#HERE WE EVALUATE OUR MODEL AGAINST THE TEST DATASET,NOT SEEN BY THE MODEL BEFORE
test_loss,test_acc= model.evaluate(test_images,test_labels)
print('Test Accuracy', test_acc)
```
